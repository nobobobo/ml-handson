```python
img_bgr = cv2.imread('asset/noboru.jpg')
print(img_bgr)
```

```python
plt.imshow(img_bgr)
plt.show()
```

## Change the color of the picture

### BGR to RGB

```python
# print shape of img_bgr
x = np.array(img_bgr)
print(x.shape)
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

### BGR to HSV

```python
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
plt.show()
```

### RGB to Gray scale

```python
img_grey = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(img_grey)
plt.show()
```

## Save converted picture

```python
cv2.imwrite('asset/hsv_noboru.jpg', img_hsv)
cv2.imwrite('asset/grey_noboru.jpg', img_grey)
```
